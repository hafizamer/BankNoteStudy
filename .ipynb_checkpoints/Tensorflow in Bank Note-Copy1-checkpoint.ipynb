```python
import pandas as pd
```

```python
data=pd.read_csv('bank_note_data.csv')
```

```python
data.head()
```

```python
import seaborn as sns
%matplotlib inline
```

```python
sns.countplot(x='Class', data=data)
```

```python
sns.pairplot(data,hue='Class')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(data.drop('Class',axis=1))
```

```python
scaled_features=scaler.transform(data.drop('Class',axis=1))
```

```python
df_feat=pd.DataFrame (scaled_features,columns=data.columns[:-1])
```

```python
df_feat.head()
```

```python
x=df_feat
```

```python
y=data['Class']
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
#egewg
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
import tensorflow as tf
```

```python
df_feat.columns
```

```python
feat_cols=[]

for col in df_feat.columns:
    feat_cols.append(tf.feature_column.numeric_column(col))
```

```python
feat_cols
```

```python
classifier=tf.estimator.DNNClassifier(hidden_units=[10,10,10],n_classes=2,feature_columns=feat_cols)
```

```python
 tf.compat.v1.estimator.inputs.pandas_input_fn
```

```python
#test
```

```python
input_func= tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_train,y=y_train,batch_size=5,shuffle=True)
```

```python
classifier.train(input_fn=input_func,steps=500)
```

```python
pred_fn=tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_test,batch_size=len(X_test),shuffle=False)
```

```python
note_predictions=list(classifier.predict(input_fn=pred_fn))
```

```python
note_predictions[0]
```

```python
final_preds=[]
for pred in note_predictions:
    final_preds.append(pred['class_ids'][0])
```

```python
final_preds
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(confusion_matrix(y_test,final_preds))
```

```python
print(classification_report(y_test,final_preds))
```

```python
model=tf.estimator.LinearClassifier(feature_columns=feat_cols,n_classes=2)
model.train(input_fn=input_func, steps=5000)
eval_input_func=tf.compat.v1.estimator.inputs.pandas_input_fn(x=X_test, y=y_test, batch_size=10,num_epochs=1,shuffle=False)
results=model.evaluate(eval_input_func)
```
